```python

import random

def print_rules():
    print("Welcome to Rock-Paper-Scissors!")
    print("Rules:")
    print("1 - Rock")
    print("2 - Paper")
    print("3 - Scissors")
    print("Enter your choice. To stop playing, enter '0'.")

def get_user_choice():
    while True:
        try:
            choice = int(input("Enter your choice: "))
            if choice not in [0, 1, 2, 3]:
                print("Please enter a valid option (0, 1, 2, or 3).")
                continue
            else:
                return choice
        except ValueError:
            print("Please enter a valid option (0, 1, 2, or 3).")

def get_choice_name(choice):
    if choice == 1:
        return "Rock"
    elif choice == 2:
        return "Paper"
    elif choice == 3:
        return "Scissors"
    else:
        return "Invalid"

def play_game():
    while True:
        print_rules()

        user_choice = get_user_choice()
        if user_choice == 0:
            print("Thanks for playing. Goodbye!")
            break

        user_choice_name = get_choice_name(user_choice)
        print("You chose:", user_choice_name)

        computer_choice = random.randint(1, 3)
        computer_choice_name = get_choice_name(computer_choice)
        print("Computer chose:", computer_choice_name)

        if user_choice == computer_choice:
            print("It's a tie!")
        elif (user_choice == 1 and computer_choice == 3) or \
             (user_choice == 2 and computer_choice == 1) or \
             (user_choice == 3 and computer_choice == 2):
            print("You win!")
        else:
            print("Computer wins!")

if __name__ == "__main__":
    play_game()
```

```python
import random

# Display rules and game options
print("Welcome to Rock-Paper-Scissors!\n")
print("Rules:\n1 - Rock\n2 - Paper\n3 - Scissors\nEnter your choice. To stop, enter '0'.")

# Define choice names for readability
choice_names = {1: "Rock", 2: "Paper", 3: "Scissors"}

# Main game loop
while True:
  # Prompt for user choice (handle invalid input directly)
  while True:
    try:
      user_choice = int(input("Enter your choice: "))
      if user_choice not in [0, 1, 2, 3]:
        print("Invalid choice. Please enter 0, 1, 2, or 3.")
        continue
      break
    except ValueError:
      print("Invalid input. Please enter a number.")

  # Exit if user chooses 0
  if user_choice == 0:
    print("Thanks for playing! Goodbye.")
    break

  # Get computer choice and choose names
  computer_choice = random.randint(1, 3)
  user_choice_name = choice_names[user_choice]
  computer_choice_name = choice_names[computer_choice]

  # Show choices and outcome
  print("\nYou chose:", user_choice_name)
  print("Computer chose:", computer_choice_name)

  if user_choice == computer_choice:
    print("It's a tie!")
  elif (user_choice == 1 and computer_choice == 3) or \
       (user_choice == 2 and computer_choice == 1) or \
       (user_choice == 3 and computer_choice == 2):
    print("You win!")
  else:
    print("Computer wins!")

  # Ask to play again and repeat if yes
  play_again = input("Play again? (Y/N): ").upper()
  if play_again != "Y":
    print("Thanks for playing! Goodbye.")
    break
```
