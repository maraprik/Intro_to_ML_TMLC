```python
import os
os.chdir=("/Users/Sarthak/Pictures")
print ( os.getcwd())

from datetime import date , timedelta

def create_file():
    fl_nm=input('Enter file name \n')
        

    try:
        with open(fl_nm,'w') as fl:
            print (" Give few lines for input, to stop enter ##")
            while True: 
                fl_txt= input('Enter txt for file\n  ')
                if len(fl_txt) > 2 and fl_txt.find('##') < 0:
                    fl_txt= fl_txt + '\n'
                    fl.write(fl_txt)
                elif len(fl_txt) > 2 and fl_txt.find('##')>= 0: 
                    fl_txt = fl_txt.replace('##','',1)
                    fl_txt= fl_txt + '\n'
                    fl.write(fl_txt)
                    break
                else:
                    break

                    
            print("File " + fl_nm +" created")
            fl.close()
    except : 
        print('File not created')

           
def copy_file():
    fl_nm_src= input('Enter  source file name \n')
    fl_nm_tar= input('Enter  target file name \n')
    list_txt=[]

    with open(fl_nm_src ,'r') as fl_read:
        lines=  fl_read.readlines()
        with open(fl_nm_tar,'w') as fl_write:
            for line in lines:
                fl_write.write(line)
    
    fl_read.close()
    fl_write.close()        
      
        
        
def dt_calc():
    dt=date.today()
    print("Today's date " + str(dt))
    
 
    dt_nu = input('Number of days from current date \n ')
    fu_dt = dt + timedelta(int(dt_nu)) 
    
    print("Asked date " + str(fu_dt))
 
      
def dt_diff():
    print ('\nPlease enter first date ')
    year = int(input('Enter a year: '))
    month = int(input('Enter a month: '))
    day = int(input('Enter a day: '))

    d1 = date(year, month, day)
    print(d1)
    
    print ('\nPlease enter second date ')
    year = int(input('Enter a year: '))
    month = int(input('Enter a month: '))
    day = int(input('Enter a day: '))

    d2 = date(year, month, day)
    print(d2)
        
    diff = (d2 -d1).days
    print(str(diff) + " days")
        
    
  
    
        
def main():
    print("File Handling ")
    print("1. create file")
    print("2. copy file")
    print("3. Date calculation")
    print("4. Date differnece")
    argu= input("options : ")
    
    if argu == "2":
        copy_file()
    elif argu == "1":
        create_file()
    elif argu == "3":
        dt_calc()
    elif argu == "4":
        dt_diff()
    else:
        print ('wait')

if __name__ == "__main__":
   main() 
```
